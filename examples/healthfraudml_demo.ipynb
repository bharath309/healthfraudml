## Step 2: Basic Claims Fraud Detection
In this section, we generate synthetic claims data, train a Machine Learning model using the unified `FraudDetector` API, and view explainable predictions.

```python
import numpy as np
from sklearn.datasets import make_classification
from healthfraudml import FraudDetector
from healthfraudml.models.supervised.random_forest import RandomForestDetector

# 1. Generate synthetic claims data (mocking claims features like codes, amounts, providers)
X, y = make_classification(
    n_samples=2000,
    n_features=15,
    n_informative=10,
    n_classes=2,
    weights=[0.96, 0.04],  # 4% fraud rate
    random_state=42
)

# Split into training and testing sets
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# 2. Train the detector
print("Training the FraudDetector using RandomForest...")
detector = FraudDetector(model=RandomForestDetector(n_estimators=100), threshold=0.4)
detector.fit(X_train, y_train)

# 3. Run predictions with explainability turned on
results = detector.predict(X_test, explain=True)

print(f"\nAnalysis Completed:")
print(f"Total claims analyzed: {len(results.predictions)}")
print(f"Flagged claims: {results.n_flagged}")
print(f"Flagged rate: {results.n_flagged / len(results.predictions):.1%}\n")

# 4. Display the top 5 flagged claims with explanations
print("TOP FLAGGED CLAIMS:")
print("=" * 40)
for case in results.flagged[:5]:
    print(f"Claim ID: {case.id}")
    print(f"Fraud Score: {case.score:.3f}")
    print(f"Explanation: {case.explanation}")
    print("-" * 40)
```

## Step 3: Evaluate Model Performance
Let's check the evaluation metrics, including Precision, Recall, and the Matthews Correlation Coefficient (MCC).

```python
metrics = detector.evaluate(X_test, y_test)
print("Evaluation Metrics:")
print(f"  Precision: {metrics.get('precision', 0):.3f}")
print(f"  Recall:    {metrics.get('recall', 0):.3f}")
print(f"  F1-Score:  {metrics.get('f1', 0):.3f}")
print(f"  MCC:       {metrics.get('mcc', 0):.3f}")
```

## Step 4: Organizational Readiness Assessment
Evaluate whether your organization is ready to adopt ML-based fraud detection. Grounded in TAM and Diffusion of Innovations theory.

```python
from healthfraudml.readiness import ReadinessAssessment

assessment = ReadinessAssessment()

# Run evaluation for a medium-sized healthcare facility
report = assessment.evaluate(
    institution_size="medium",
    annual_claims_volume=120000,
    existing_fraud_detection="manual",
    it_staff_count=5,
    annual_fraud_budget=45000,
    has_data_science_team=False,
    data_digitized=True,
    has_labeled_fraud_data=True,
    leadership_support=True,
    staff_tech_comfort=3,
    regulatory_requirements=True
)

print("=" * 60)
print("ORGANIZATIONAL READINESS REPORT")
print("=" * 60)
print(report.summary())

print("\nPrioritized Recommendations:")
for i, rec in enumerate(report.recommendations, 1):
    print(f"  {i}. {rec}")

print("\nImplementation Roadmap:")
for phase in report.implementation_roadmap:
    print(f"  Phase {phase['phase']} (Months {phase['months']}): {phase['focus']}")
    for action in phase['actions']:
        print(f"    - {action}")
```
